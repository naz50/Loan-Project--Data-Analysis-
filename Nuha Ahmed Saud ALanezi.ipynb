## Apply non Return funcation 

```python
def get_Gender(Gender,Loan_ID,Married,Dependents,Education,Loan_Status):
    if  Gender=='Male':
        print( 'This your in information Mr.','Loan_ID is ',Loan_ID,'Married Status is ',Married,'Dependents is',Dependents,'Education is',Education,'Loan_Status is',Loan_Status)
    elif Gender =='Female':
         print( 'This your in information Ms.','Loan_ID is ',Loan_ID,'Married Status is ',Married,'Dependents is',Dependents,'Education is',Education,'Loan_Status is',Loan_Status)
    else:
         print( 'UnKnown')
```

## Apply Return funcation 

```python
def get_Loan_Amount_Term(Loan_Amount_Term):
    if 480>= Loan_Amount_Term > 342 :
        return 'high'
    elif 342 >= Loan_Amount_Term > 36:
        return 'medium'
    else:
        return 'low'
    
```

## make LabelEncoder to more than two types & get_dummies to two types

```python
data['Gender'] = pd.get_dummies(data['Gender'], drop_first=True) 
```

```python
data['Married'] = pd.get_dummies(data['Married'], drop_first=True)
```

```python
data['Education']=pd.get_dummies(data['Education'], drop_first=True)
```

```python
data['Loan_Status']=pd.get_dummies(data['Loan_Status'], drop_first=True)
```
